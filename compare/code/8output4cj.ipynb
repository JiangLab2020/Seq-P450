```python
import numpy as np
import pandas as pd
```

```python
ways = ["our", "esp", "enzypick", "puepp"]
datas = [
    "Arabidopsis_thaliana",
    "Glycine_max",
    "Erigeron_breviscapus",
    "Zea_mays",
    "sheetall",
]
topk = 0.05
```

```python
for data in datas:
    for way in ways:
        if way == "our":
            exec(f"df_{way}_{data} = pd.read_csv(f'../input/{data}.csv')")
        else:
            exec(f"df_{way}_{data} = pd.read_csv(f'../out4cj/{way}_{data}_{topk}.csv')")
```

```python
df_esp_Arabidopsis_thaliana = df_esp_Arabidopsis_thaliana.rename(
    columns={"scores": "ESP_score"}
)

# -----------------------------------Arabidopsis_thaliana-----------------------------------
df_our_Arabidopsis_thaliana["substrate_cc"] = (
    df_our_Arabidopsis_thaliana["substrate"] + df_our_Arabidopsis_thaliana["cc"]
)
result_grouped = df_our_Arabidopsis_thaliana.groupby("substrate_cc")

filtered_rows = df_our_Arabidopsis_thaliana[
    (df_our_Arabidopsis_thaliana["if_right"])
    & (df_our_Arabidopsis_thaliana["Binding"] == 1)
]["substrate_cc"].tolist()
df_our_Arabidopsis_thaliana = df_our_Arabidopsis_thaliana[
    df_our_Arabidopsis_thaliana["substrate_cc"].isin(filtered_rows)
]
# -----------------------------------Arabidopsis_thaliana-----------------------------------

df_Arabidopsis_thaliana = pd.merge(
    df_our_Arabidopsis_thaliana,
    df_esp_Arabidopsis_thaliana,
    on=["enzyme", "substrate"],
    how="inner",
)
df_enzypick_Arabidopsis_thaliana = df_enzypick_Arabidopsis_thaliana.rename(
    columns={"DL Model Score": "enzypick_score"}
)
df_Arabidopsis_thaliana = pd.merge(
    df_Arabidopsis_thaliana,
    df_enzypick_Arabidopsis_thaliana,
    on=["Sequence", "substrate"],
    how="left",
)
df_puepp_Arabidopsis_thaliana = df_puepp_Arabidopsis_thaliana.rename(
    columns={"score": "PUEPP_score"}
)
df_puepp_Arabidopsis_thaliana = df_puepp_Arabidopsis_thaliana.drop(
    columns=["ESP_p450_score", "Gnina_score"]
)
df_Arabidopsis_thaliana = pd.merge(
    df_Arabidopsis_thaliana,
    df_puepp_Arabidopsis_thaliana,
    on=["enzyme", "substrate", "Binding", "if_right", "cc"],
    how="left",
)

df_Arabidopsis_thaliana["P450Mining_score"] = (
    0.3 * df_Arabidopsis_thaliana["ESP_p450_score"]
    + 0.7 * df_Arabidopsis_thaliana["Gnina_score"]
)
df_Arabidopsis_thaliana[
    [
        "enzyme",
        "substrate",
        "Binding",
        "if_right",
        "cc",
        "P450Mining_score",
        "ESP_p450_score",
        "Gnina_score",
        "ESP_score",
        "enzypick_score",
        "PUEPP_score",
    ]
].to_csv("../out4cj/final/Arabidopsis_thaliana.csv")
```

```python
df_esp_Glycine_max = df_esp_Glycine_max.rename(columns={"scores": "ESP_score"})


# -----------------------------------Arabidopsis_thaliana-----------------------------------
df_our_Glycine_max["substrate_cc"] = (
    df_our_Glycine_max["substrate"] + df_our_Glycine_max["cc"]
)
result_grouped = df_our_Glycine_max.groupby("substrate_cc")

filtered_rows = df_our_Glycine_max[
    (df_our_Glycine_max["if_right"]) & (df_our_Glycine_max["Binding"] == 1)
]["substrate_cc"].tolist()
df_our_Glycine_max = df_our_Glycine_max[
    df_our_Glycine_max["substrate_cc"].isin(filtered_rows)
]
# -----------------------------------Arabidopsis_thaliana-----------------------------------


df_Glycine_max = pd.merge(
    df_our_Glycine_max, df_esp_Glycine_max, on=["enzyme", "substrate"], how="left"
)
df_enzypick_Glycine_max = df_enzypick_Glycine_max.rename(
    columns={"DL Model Score": "enzypick_score"}
)
df_Glycine_max = pd.merge(
    df_Glycine_max, df_enzypick_Glycine_max, on=["Sequence", "substrate"], how="left"
)
df_puepp_Glycine_max = df_puepp_Glycine_max.rename(columns={"score": "PUEPP_score"})
df_puepp_Glycine_max = df_puepp_Glycine_max.drop(
    columns=["ESP_p450_score", "Gnina_score"]
)
df_Glycine_max = pd.merge(
    df_Glycine_max,
    df_puepp_Glycine_max,
    on=["enzyme", "substrate", "Binding", "if_right", "cc"],
    how="left",
)


df_Glycine_max["P450Mining_score"] = (
    0.3 * df_Glycine_max["ESP_p450_score"] + 0.7 * df_Glycine_max["Gnina_score"]
)
df_Glycine_max = df_Glycine_max.drop_duplicates(
    subset=["enzyme", "substrate"], keep="first"
)
df_Glycine_max[
    [
        "enzyme",
        "substrate",
        "Binding",
        "if_right",
        "cc",
        "P450Mining_score",
        "ESP_p450_score",
        "Gnina_score",
        "ESP_score",
        "enzypick_score",
        "PUEPP_score",
    ]
].to_csv("../out4cj/final/Glycine_max.csv")
```
